```python
# ==========================================
# STEP 1: MODELING FUNCTIONS
# ==========================================

import numpy as np
import matplotlib.pyplot as plt

# The Answer to the Ultimate Question of Life, the Universe, and Everything: 42
np.random.seed(42)

M = 1000

# Generating aleatory uniform distribution
x = np.random.uniform(low=0.0, high=1.0, size=(M, 1))

# Generating normal Gaussian distribution (sqrt(1) = 1)
w = np.random.normal(loc=0.0, scale=1.0, size=(M, 1))

# Noisy Equation
y = 1.2 + 2.3 * x + 10 * w

# Pure Equation
true_y = 1.2 + 2.3 * x
```

```python
# ==========================================
# STEP 2: PLOTING (Paper/IEEE Format)
# ==========================================

plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Times New Roman', 'DejaVu Serif'],
    'font.size': 10,
    'axes.labelsize': 10,
    'axes.titlesize': 11,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'figure.dpi': 300,
    'savefig.dpi': 300,
})

# Colors
function_color = '#4C72B0'   # Blue
point_color = '#DD8452'  # Orange

plt.figure(figsize=(8, 5))

# Ploting Noisy Sample as points
plt.scatter(x, y, color=point_color, alpha=1, s=10, label='Noisy Sample')

# Ploting Pure Function as line
plt.plot(x, true_y, color=function_color, alpha=0.6, label='Function ($y = 2.76x$)', linewidth=2)

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')

plt.grid(axis='both', linestyle=':', alpha=0.6)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# ==========================================
# STEP 1: MODELING FUNCTIONS
# ==========================================

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

M = 10000

# Generating aleatory uniform distribution
x = np.random.uniform(0, 1, size=(M, 1))
y = np.random.uniform(0, 1, size=(M, 1))

z = x + y
```

```python
# ==========================================
# STEP 2: PLOTING (Paper/IEEE Format)
# ==========================================

# 1. Global Matplotlib IEEE Configuration
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Times New Roman', 'DejaVu Serif'], # Font
    'font.size': 10,           # Font Size
    'axes.labelsize': 10,      # Labels (X e Y)
    'axes.titlesize': 11,      # Graphic Titles
    'xtick.labelsize': 9,      # Values X
    'ytick.labelsize': 9,      # Values Y
    'figure.dpi': 300,
    'savefig.dpi': 300,
    'hatch.linewidth': 0.4,
})

# Colors
orange = '#DD8452'  # Orange

plt.figure(figsize=(8, 5))

plt.hist(z, bins=50, density=True, color=orange, edgecolor='black', alpha=0.8, hatch='xxxx')
plt.title('Histograma Normalizado de $z = x + y$')
plt.xlabel('Valores de z')
plt.ylabel('Densidade de Probabilidade')

plt.tight_layout()
plt.show()
```

```python
# ==========================================
# STEP 1: MODELING FUNCTIONS
# ==========================================

import numpy as np
import matplotlib.pyplot as plt

# Generating uniformy spaced values from -10 to 11
# 11+0.25, because as seen in Ref [5], stop does not include the 11 value
x1_base = np.arange(start=-10, stop=(11+0.25), step=0.25)
x2_base = np.arange(start=-10, stop=(11+0.25), step=0.25)
x1, x2 = np.meshgrid(x1_base, x2_base)

y = (x1**2) + (x2**2)
```

```python
# ==========================================
# STEP 2: PLOTING (Paper/IEEE Format)
# ==========================================

# 1. Global Matplotlib IEEE Configuration
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Times New Roman', 'DejaVu Serif'], # Font
    'font.size': 10,           # Font Size
    'axes.labelsize': 10,      # Labels (X e Y)
    'axes.titlesize': 11,      # Graphic Titles
    'xtick.labelsize': 9,      # Values X
    'ytick.labelsize': 9,      # Values Y
    'figure.dpi': 300,
    'savefig.dpi': 300,
    'hatch.linewidth': 0.4,
})

# Colors
orange = '#DD8452'  # Orange

fig = plt.figure(figsize=(8, 5))

# Add 3D axis
ax = fig.add_subplot(111, projection='3d')

# Plot surface with 'viridis' map color
surf = ax.plot_surface(x1, x2, y, cmap='viridis', edgecolor='none', alpha=0.9)

ax.set_title('Superfície $y = x_1^2 + x_2^2$')
ax.set_xlabel('Eixo $x_1$')
ax.set_ylabel('Eixo $x_2$')
ax.set_zlabel('Eixo $y$')

fig.colorbar(surf, ax=ax, shrink=0.4, aspect=10, label='Valor de $y$')

plt.tight_layout()
plt.show()
```
