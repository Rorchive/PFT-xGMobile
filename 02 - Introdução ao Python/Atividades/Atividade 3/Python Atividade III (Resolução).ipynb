**Seção 1 — Operações Básicas e Variáveis**

**Exercício 1:** crie três variáveis numéricas e exiba a soma delas.

````python
import math

def somar(n1, n2, n3):
  return n1 + n2 + n3

def triangular(number):
  # Um número n é triangular se 8n + 1 é um quadrado perfeito
  if number < 0:
    return False
  raiz = math.sqrt(8 * number + 1)
  return raiz == int(raiz) and int(raiz) % 2 != 0
````

````python
number1 = 3
number2 = 4
number3 = 6

# Soma
soma = somar(number1, number2, number3)
print(f"{number1} + {number2} + {number3} = {soma}")

# Testando a função triangular com o número 3
number = 3
is_triangular = triangular(number)

if is_triangular:
  print(f"O número {number} é triangular.")
else:
  print(f"O número {number} não é triangular.")
````

**Exercício 7**

Crie uma função que recebe um número e retorna seu fatorial.

````python
numero = int(input("Digite um número: "))

def fatorial(numero):
  if numero == 0 or numero == 1:
    return 1
  else:
    return numero * fatorial(numero - 1)

print(f"O fatorial do número {numero} é: {fatorial(numero)}")
````

**Seção 11 — Raise**

**Exercício 18**

Crie uma função chamada `validar_senha(senha)` que deve verificar:

* A senha deve ter pelo menos 8 caracteres
* Deve conter pelo menos um número
* Deve conter pelo menos uma letra maiúscula

Se qualquer uma dessas regras não for cumprida, a função deve lançar uma exceção personalizada chamada:

```
class SenhaInvalidaError(Exception):
    pass
```



A função deve retornar `"Senha válida!"` se tudo estiver correto.

````python
class SenhaInvalidaError(Exception):
  pass

def validar_senha(senha):
  if len(senha) < 8:
    raise SenhaInvalidaError("A senha deve ter pelo menos 8 caracteres.")

  if not any(char.isdigit() for char in senha):
    raise SenhaInvalidaError("A senha deve conter pelo menos um número.")

  if not any(char.isupper() for char in senha):
    raise SenhaInvalidaError("A senha deve conter pelo menos uma letra maiúscula.")

  return "Senha válida!"

senha = input("Digite uma senha: ")

try:
  print(validar_senha(senha))
except SenhaInvalidaError as e:
  print(e)
````

**Seção 12 — Desafio**

**Exercício 19**

Crie um programa que:

1. Peça nome e idade de 5 pessoas

2. Armazene em uma lista de dicionários

3. Informe quem é a pessoa mais velha

````python
pessoas = []

for i in range(5):
    nome = input(f"Digite o nome da pessoa {i + 1}: ")

    while True:
        try:
            idade = int(input(f"Digite a idade de {nome}: "))
            break
        except ValueError:
            print("Digite apenas números inteiros para a idade!")

    pessoas.append({"nome": nome, "idade": idade})

pessoa_mais_velha = max(pessoas, key=lambda x: x["idade"])
print(f"A pessoa mais velha é {pessoa_mais_velha["nome"]} com {pessoa_mais_velha["idade"]} anos.")
````
