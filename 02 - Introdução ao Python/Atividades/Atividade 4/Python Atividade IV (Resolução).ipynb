```python
# ============================================================
# EXERCÍCIOS SOBRE map, filter, lambda, pandas e numpy
# ============================================================

# ============================
# 1) MAP + LAMBDA
# ============================
# Usando a lista abaixo, use map + lambda para multiplicar cada valor por 2.
# Salve o resultado em uma lista chamada "resultado1".
numeros = [1, 3, 5, 7, 9]

resultado1 = list(map(lambda x: x * 2, numeros))

print(resultado1)


# ============================
# 2) FILTER + LAMBDA
# ============================
# Usando filter + lambda, filtre apenas os números pares da lista abaixo.
# Salve o resultado em "resultado2".
valores = [10, 15, 22, 33, 40, 55]

resultado2 = list(filter(lambda x: x % 2 == 0, valores))

print(resultado2)


# ============================
# 3) MAP + FILTER + LAMBDA
# ============================
# Primeiro filtre os nomes com mais de 5 letras usando filter + lambda,
# depois converta todos para letras maiúsculas usando map + lambda.
# Salve o resultado final em "resultado3".
nomes = ["Ana", "Roberto", "Lara", "Carolina", "Pedro"]

resultado3 = list(map(lambda x: x.upper(), filter(lambda x: len(x) > 5, nomes)))

print(resultado3)

# ============================
# 4) REDUÇÃO DE PREÇOS COM map
# ============================
# A lista "precos" representa valores de produtos.
# Use map + lambda para aplicar 10% de desconto em cada item.
# Salve o resultado em "precos_com_desconto".
precos = [120.0, 50.0, 89.9, 230.0]

precos_com_desconto = list(map(lambda x: x * 0.9, precos))

print(precos_com_desconto)


# ============================
# 5) NUMPY BÁSICO
# ============================
# Crie um array numpy a partir da lista abaixo.
# Depois multiplique todo o array por 3.
# Salve o array final em "array3x".
import numpy as np
lista_np = [2, 4, 6, 8]

array_numpy = np.array(lista_np)
array3x = array_numpy * 3

print(array3x)


# ============================
# 6) NUMPY: ESTATÍSTICAS
# ============================
# Usando numpy, calcule:
# - média
# - desvio padrão
# do array abaixo.
# Salve os resultados em media_np e desvio_np.
array_est = np.array([10, 20, 30, 40, 50])

media_np = np.mean(array_est)
desvio_np = np.std(array_est)

print(media_np)
print(desvio_np)

# ============================
# 7) PANDAS: CRIAÇÃO DE DATAFRAME
# ============================
# Crie um DataFrame pandas com duas colunas:
# "nome": ["Ana", "Bruno", "Carlos"]
# "idade": [23, 31, 19]
# Salve o dataframe em df.
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carlos"],
    "idade": [23, 31, 19]
}

df = pd.DataFrame(dados)

print(df)
```
